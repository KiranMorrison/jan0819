```python
id__ = [354545, 354547]
```

```python
beachname = []
beachid = []
date = []
uvlevel = []
time = []
for i in id__:
    url = 'http://datapoint.metoffice.gov.uk/public/data/val/wxfcs/all/json/{}?res=3hourly&key=eadde26b-701c-4d9a-8849-17d749bb9380'.format(i)
    req = request.Request(url,)
    d = request.urlopen(req)
    data = d.read().decode('utf-8')
    places = [json.loads(data)]
    for dic in places:
        beachname.append(dic.get('SiteRep').get('DV').get('Location').get('name'))
        beachid.append(dic.get('SiteRep').get('DV').get('Location').get('i'))
        for r in range(0,5):
            date.append(dic.get('SiteRep').get('DV').get('Location').get('Period')[r].get('value'))
            for x in range(0,len(dic.get('SiteRep').get('DV').get('Location').get('Period')[r].get('Rep'))):
                uvlevel.append(dic.get('SiteRep').get('DV').get('Location').get('Period')[r].get('Rep')[x].get('U'))
                time.append(dic.get('SiteRep').get('DV').get('Location').get('Period')[r].get('Rep')[x].get('$'))
                
```

```python
keys = ['beachname', 'beachid', 'date', 'uvlevel', 'time']
values = [beachname, beachid, date, uvlevel, time]
dictionary = dict(zip(keys, values))
```
